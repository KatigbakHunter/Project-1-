```python
import pandas as pd
import os 
import matplotlib.pyplot as plt
```

```python
path = "/Users/katigbakv/Desktop/adult-depression-lghc-indicator-24.csv"
adult_df = pd.read_csv(path)
adult_df.head(5)
```

```python
race_data = adult_df[adult_df['Strata'] == 'Race-Ethnicity']
race_data.head()
```

```python
white = race_data[race_data['Strata Name'] == 'White']
white
```

```python
Other = race_data[race_data['Strata Name'] == 'Other']
Other
```

```python
Black = race_data[race_data['Strata Name'] == 'Black']
Black
```

```python
Asian = race_data[race_data['Strata Name'] == 'Asian/Pacific Islander']
Asian
```

```python
Hispanic = race_data[race_data['Strata Name'] == 'Hispanic']
Hispanic
```

```python
plt.figure(figsize=(12, 8))

plt.plot(white['Year'], white['Percent'], marker='o', linestyle='-', linewidth=2, label='White')
plt.plot(Other['Year'], Other['Percent'], marker='o', linestyle='-', linewidth=2, label='Other')
plt.plot(Black['Year'], Black['Percent'], marker='o', linestyle='-', linewidth=2, label='Black')
plt.plot(Asian['Year'], Asian['Percent'], marker='o', linestyle='-', linewidth=2, label='Asian/Pacific Islander')
plt.plot(Hispanic['Year'], Hispanic['Percent'], marker='o', linestyle='-', linewidth=2, label='Hispanic')


plt.xlabel("Year", fontsize=12)
plt.ylabel("Percent", fontsize=12)
plt.title("Trend of Depression Cases by Year for Each Ethnicity", fontsize=14)
plt.xticks(race_data['Year'].unique(), rotation=45)


plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))


bar_width = 0.15
x = race_data['Year'].unique()  


positions_white = x - 2 * bar_width
positions_other = x - bar_width
positions_black = x
positions_asian = x + bar_width
positions_hispanic = x + 2 * bar_width

plt.bar(positions_white, white['Frequency'], width=bar_width, alpha=0.5, label='White', color='blue')
plt.bar(positions_other, Other['Frequency'], width=bar_width, alpha=0.5, label='Other', color='orange')
plt.bar(positions_black, Black['Frequency'], width=bar_width, alpha=0.5, label='Black', color='green')
plt.bar(positions_asian, Asian['Frequency'], width=bar_width, alpha=0.5, label='Asian/Pacific Islander', color='purple')
plt.bar(positions_hispanic, Hispanic['Frequency'], width=bar_width, alpha=0.5, label='Hispanic', color='red')


plt.xlabel("Year", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Comparison of Depression Cases by Ethnicity and Year", fontsize=14)
plt.xticks(x, rotation=45) 


plt.legend()
plt.tight_layout()
plt.show()
```
