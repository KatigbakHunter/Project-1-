```python
import pandas as pd
import os 
import matplotlib.pyplot as plt
```

```python
path = "/Users/katigbakv/Desktop/adult-depression-lghc-indicator-24.csv"
adult_df = pd.read_csv(path)
adult_df.head(5)
```

```python
education_data = adult_df[adult_df['Strata Name'].str.contains("High School|College|Diploma", na=False)]
education_data.head(10)
```

```python

education_levels = education_data['Strata Name'].unique()
plt.figure(figsize=(12, 8))


for level in education_levels:
    group_data = education_data[education_data['Strata Name'] == level]
    plt.plot(group_data['Year'], group_data['Percent'], marker='o', linestyle='-', linewidth=2, label=level)


plt.xlabel("Year", fontsize=12)
plt.ylabel("Percent", fontsize=12)
plt.title("Trend of Depression Percentages by Education Level Over the Years", fontsize=14)
plt.xticks(education_data['Year'].unique(), rotation=45)
plt.grid()
plt.legend(title="Education Levels")
plt.tight_layout()
plt.show()
```
