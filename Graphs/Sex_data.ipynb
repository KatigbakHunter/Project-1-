```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
path = "/Users/katigbakv/Desktop/adult-depression-lghc-indicator-24.csv"
adult_df = pd.read_csv(path)
adult_df.head(15)
```

```python
gender_data = adult_df[adult_df['Strata'] == 'Sex']
gender_data
```

```python
female_data = gender_data[gender_data['Strata Name'] == 'Female']
female_data
```

```python
male_data = gender_data[gender_data['Strata Name'] == 'Male']
male_data
```

```python
plt.figure(figsize=(10, 6))
plt.bar(male_data['Year'], male_data['Percent'], width=0.6)

plt.xlabel("Year", fontsize=12)
plt.ylabel("Percent", fontsize=12)
plt.title("Percentage of Depression Cases (Male) by Year", fontsize=14)
plt.xticks(male_data['Year'], rotation=45)  
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(female_data['Year'], female_data['Percent'], width=0.6, color='pink')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Percent", fontsize=12)
plt.title("Percentage of Depression Cases (Female) by Year", fontsize=14)
plt.xticks(female_data['Year'], rotation=45)  
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))


bar_width = 0.4


plt.bar(male_data['Year'] - bar_width / 2, male_data['Percent'], width=bar_width, alpha=0.5, color='blue', label='Male')

plt.bar(female_data['Year'] + bar_width / 2, female_data['Percent'], width=bar_width, alpha=0.5, color='pink', label='Female')


plt.xlabel("Year", fontsize=12)
plt.ylabel("Percent", fontsize=12)
plt.title("Percentage of Depression Cases by Year (Male vs Female)", fontsize=14)
plt.xticks(male_data['Year'], rotation=45) 


plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(male_data['Year'], male_data['Percent'], marker='o', linestyle='-', linewidth=2)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Percent", fontsize=12)
plt.title("Trend of Depression Cases (Male) by Year", fontsize=14)
plt.xticks(male_data['Year'], rotation=45)  
plt.tight_layout()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(female_data['Year'], female_data['Percent'], marker='o', linestyle='-', linewidth=2, color='pink')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Percent", fontsize=12)
plt.title("Trend of Depression Cases (Female) by Year", fontsize=14)
plt.xticks(female_data['Year'], rotation=45)  
plt.tight_layout()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(male_data['Year'], male_data['Percent'], marker='o', linestyle='-', linewidth=2, color='blue', label='Male')
plt.plot(female_data['Year'], female_data['Percent'], marker='o', linestyle='-', linewidth=2, color='pink', label='Female')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Percent", fontsize=12)
plt.title("Trend of Depression Cases by Year (Male vs Female)", fontsize=14)
plt.xticks(male_data['Year'], rotation=45)
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```
