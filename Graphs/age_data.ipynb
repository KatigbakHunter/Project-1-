```python
import pandas as pd
import os 
import matplotlib.pyplot as plt
```

```python
path = "/Users/katigbakv/Desktop/adult-depression-lghc-indicator-24.csv"
adult_df = pd.read_csv(path)
adult_df.head(5)
```

```python
age_data = adult_df[adult_df['Strata'] == 'Age']
age_data
```

```python
age_groups = age_data['Strata Name'].unique()

plt.figure(figsize=(12, 8))


for age_group in age_groups:
    group_data = age_data[age_data['Strata Name'] == age_group]
    plt.plot(group_data['Year'], group_data['Percent'], marker='o', linestyle='-', linewidth=2, label=age_group)

plt.xlabel("Year", fontsize=12)
plt.ylabel("Percent", fontsize=12)
plt.title("Trend of Depression Percentages by Age Group Over the Years", fontsize=14)
plt.xticks(age_data['Year'].unique(), rotation=45)
plt.grid()
plt.legend(title="Age Groups")
plt.tight_layout()
plt.show()
```
